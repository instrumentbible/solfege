```python
#load in data

import json
import numpy as np

#function that counts the total number of datapoints 
def count_data(categories):
    total = 0
    for cat in categories:
        with open('../data/' + cat + '.json') as f:
            data = json.load(f)
            total += len(data)
    return total

categories = ['do','re','mi','fa','so','la'] #categories used for classification


dataSize = count_data(categories) #find data size
nFeatures = 42 #specify feature size (21 for x coordinates, 21 for y coordinates)

X = np.zeros((dataSize,nFeatures))
Y = np.zeros((dataSize,1))

counter = 0
for i in range(len(categories)):
    with open('../data/' + categories[i] + '.json') as f:
        data = json.load(f)
    
    for obj in data:
        x_cords = obj['normal']['x']
        y_cords = obj['normal']['y']
        
        for j in range(len(x_cords)):
            X[counter,j] = x_cords[j]
            
        for j in range(len(y_cords)):
            num = j + len(x_cords) 
            X[counter,num] = y_cords[j]
            
        Y[counter] = i
        counter += 1
```

```python
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split

Y = to_categorical(Y)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)

```

```python
from keras.models import Sequential
from keras.layers import Dense, InputLayer, Dropout, Flatten, Activation, Input

model = Sequential()
model.add(Input(shape=(42,)))
model.add(Dense(100, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(100, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(100, activation='relu'))

model.add(Dense(len(categories), activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
import matplotlib.pyplot as plt    # for plotting the images
%matplotlib inline

history = model.fit(X_train, Y_train, epochs=100, validation_data=(X_test, Y_test))

# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
from mlxtend.plotting import plot_confusion_matrix
import matplotlib.pyplot as plt

Y_pred = model.predict(X_test)

matrix = confusion_matrix(Y_test.argmax(axis=1), Y_pred.argmax(axis=1))

fig, ax = plot_confusion_matrix(conf_mat=matrix,class_names=categories,colorbar=True)

plt.show()
```
